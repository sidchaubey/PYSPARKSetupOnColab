```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName("readtextfile")
```

```
sc = SparkContext.getOrCreate(conf=conf)
rdd = sc.textFile("/FileStore/tables/map.txt")
rdd.collect()
```

```
rdd1 = rdd.map(lambda x: x.split()).flatMap(lambda x: x)
rdd1.countByValue()
```

```
rdd1.count()
```

```
rdd2 = rdd1.distinct()
rdd2.collect()
rdd2.countByValue()
```

```
import numpy as np

rdd3 = rdd1.map(lambda x: (x, np.random.randint(1, 100)))
```

```
# Finding max value per key
rdd4 = rdd3.groupByKey().mapValues(list)
```

```
rdd3.groupByKey().mapValues(list).map(lambda x: (x[0], max(x[1]))).collect()
```

```
# Reduce By Keys
# rdd4.collect() 
rdd3.map(lambda x: x).reduceByKey(lambda x,y: x+y).collect()
```

```
rdd4.collect()
```

```
# Average
rdd5 = rdd4.map(lambda x: (x[0], sum(x[1])/len(x[1])))
rdd5.collect()
```

```
even_list = rdd5.filter(lambda x: True if x[1]%2==0 else False).collect()

odd_list = rdd5.filter(lambda x: True if x[1]%2!=0 else False).collect()
```

```
odd_list
```
